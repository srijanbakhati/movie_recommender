```python
import ast
```

```python
def fetch_data(text):
    ret=[]
    for x in ast.literal_eval(text):
        ret.append(x['name'])
    return ret
```

```python
# data
```

```python
data.loc[:,'genres']=data['genres'].apply(fetch_data)
```

```python
# data
```

```python
data['keywords'][0]
```

```python
data.loc[:,'keywords']=data['keywords'].apply(fetch_data)
```

```python
data['cast'][0]
```

```python
def fetch_cast(text):
    cast=[]
    n=0
    for x in ast.literal_eval(text):
        if n<3:
            cast.append(x['name'])
            n+=1
    return cast
```

```python
data['cast']=data['cast'].apply(fetch_cast)
```

```python
data['crew'][0]
```

```python
def fetch_crew(text):
    crew=[]
    for x in ast.literal_eval(text):
        if x['job']=='Director':
            crew.append(x['name'])
            break
    return crew
```

```python
data['crew']=data['crew'].apply(fetch_crew)
```

```python
data.head()
```

```python
def join_lower(text):
    word=[]
    for x in text:
        word.append(x.replace(' ','').lower())
    return word
        
```

```python
data.loc[:,'genres']=data['genres'].apply(join_lower)
```

```python
data.loc[:,'cast']=data['cast'].apply(join_lower)
```

```python
data.loc[:,'crew']=data['crew'].apply(join_lower)
```

```python
data.loc[:,'keywords']=data['keywords'].apply(join_lower)
```

```python
data.head()
```

```python
data['overview']=data['overview'].str.split()
```

```python
data['overview']=data['overview'].apply(join_lower)
```

```python
data
```

```python
data['tags']=data['genres']+data['cast']+data['crew']+data['keywords']+data['overview']
```

```python
data
```

```python
data[['title', 'tags']].to_csv('clean_movie.csv', index=False)
```

# Removing stopwords

```python
from nltk.corpus import stopwords
```

```python
stop_words=set(stopwords.words('english'))
```

```python
stop_words
```

```python
df['tags'][0]
```

```python
df['tags']=df['tags'].apply(lambda x:[word for word in ast.literal_eval(x) if word not in stop_words])
```

```python
# df['tags'][0]
```

# stemming

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
df['tags']=df['tags'].apply(lambda x:[stemmer.stem(word) for word in x])
```

```python
df
```

# Vectorizing and cosine similarity

```python
df['tags']=df['tags'].apply(lambda x:' '.join(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vc=TfidfVectorizer(max_features=3000)
```

```python
vector=vc.fit_transform(df['tags']).toarray()
```

```python
vector.shape
```

```python
# from sklearn.pipeline import Pipeline
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# pipeline=Pipeline([
#     ('Tfidf',TfidfVectorizer(max_features=3000)),
#     ('similarity',cosine_similarity()),
# ])
#  we cannot use cosine_similarity in pipeline directly because only transformer or estimator is used in pipeline while cosine_similarity is function
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity.shape
```

```python
df[df['title']=='Spectre'].index[0]
```

```python
list(enumerate(similarity[3]))
```

```python
similar=sorted(list(enumerate(similarity[1])),reverse=True,key=lambda x:x[1])
similar
```

```python
def recommend(movie):
    index=df[df['title']==movie].index[0]
    similar=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in similar[1:6]:
        print(df.iloc[i[0]].title)
    
```

```python
recommend('Undiscovered')
```
